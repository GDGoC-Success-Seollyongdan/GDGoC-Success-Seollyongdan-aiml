```python
df1 = df1[df1["기준_년분기_코드"]==20242]
df1.fillna(0, inplace=True)
```

```python
df1['서비스_업종_코드_명'].unique()
```

```python
mart_categories = ['청과상', '수산물판매', '육류판매', '미곡판매', '편의점', '슈퍼마켓']
cafe_categories = ['커피-음료', '제과점']
restaurant_categories = ['반찬가게', '분식전문점', '치킨전문점', '패스트푸드점', '양식음식점', '일식음식점', '중식음식점', '한식음식점']

# 자치구별 업종 카테고리별 점포 수 계산
df1["카테고리"] = df1["서비스_업종_코드_명"].apply(
    lambda x: "마트" if x in mart_categories else
              "카페" if x in cafe_categories else
              "식당" if x in restaurant_categories else None
)

# NaN 값 제거 (카테고리에 포함되지 않은 것 제거)
df_filtered = df1.dropna(subset=["카테고리"])[["자치구_코드_명", "서비스_업종_코드_명", "카테고리", "점포_수"]]

# 자치구별, 카테고리별 점포 수 합계 계산
df_grouped = df_filtered.groupby(["자치구_코드_명", "카테고리"])["점포_수"].sum().unstack().reset_index()

df = df.merge(df_grouped, left_on="자치구", right_on="자치구_코드_명", how="left")
df = df.drop(columns="자치구_코드_명")
```
